```python
AHDI_df.columns = list(map( lambda x : int(x/10) * 10, AHDI_df.columns.astype('int')))
AHDI_df
```

```python
df = pd.read_csv('../data/iq_data.csv')
df['Year'] = (df['Year'] // 10)
df['Year'].astype('int', )
df['Year'] = df['Year'] * 10

df.columns = ['Entity', 'Year', 'Average IQ Change', 'Fullscale_IQ_Change']

df = df[df['Fullscale_IQ_Change'].notnull()]
df.drop('Average IQ Change', inplace=True, axis=1)
df
```

```python
america = north_america + south_america
continent_names = ['Europe', 'Asia', 'Africa', 'Oceania', 'America']
continents = [europe, asia, africa, oceania, america]

final_df = pd.DataFrame(data =[], index = AHDI_df.columns)

def clean_and_mean(x):
    y = x.dropna(axis = 0)
    continent_mean.append(y.mean())
    
for i in range(5) :
    continent_name = continent_names[i]
    continent_idx = continents[i]
    
    continent_AHDI_df = AHDI_df.loc[continent_idx, :].T.copy()
    continent_mean = []
    continent_AHDI_df.apply(lambda x: clean_and_mean(x), axis=1)
    final_df[continent_name] = continent_mean
    
final_df
```

```python
# Index Entity AHDI
cont_AHDI_df = pd.DataFrame(data=[], columns = ['Year', 'Entity', 'AHDI'] )
for continent in final_df.columns :
    for i in range(len(final_df.loc[:, continent].values)) :
        year = final_df.index.values[i]
        value = final_df.loc[:, continent].values[i]
#         print(value)
        cont_AHDI_df = cont_AHDI_df.append({'Year': year, 'Entity': continent, 'AHDI': value},  ignore_index=True)
cont_AHDI_df
```

```python
!pip install pandasql
from pandasql import sqldf
```

```python
output = sqldf('''
SELECT df.Year, df.Entity, df.Fullscale_IQ_Change, cont_AHDI_df.AHDI 
FROM cont_AHDI_df
JOIN df
ON df.Year = cont_AHDI_df.Year
AND df.Entity = cont_AHDI_df.Entity
''')
columns = ['Decade', 'Continent', 'Fullscale_IQ_Change', 'Schooling_Index']
output.columns = columns 
output
```

```python
america_df = output[output['Continent'] == 'America']
asia_df = output[output['Continent'] == 'Asia']
africa_df = output[output['Continent'] == 'Africa']
oceania_df = output[output['Continent'] == 'Oceania']
eur_df = output[output['Continent'] == 'Europe']

america_df['Decade'] = (america_df['Decade'] - america_df.iloc[0,0])/10
asia_df['Decade'] = (asia_df['Decade'] - asia_df.iloc[0,0])/10
africa_df['Decade'] = (africa_df['Decade'] - africa_df.iloc[0,0])/10
oceania_df['Decade'] = (oceania_df['Decade'] - oceania_df.iloc[0,0])/10
eur_df['Decade'] = (eur_df['Decade'] - eur_df.iloc[0,0])/10

output = america_df.copy()
output = output.append(asia_df, ignore_index = True)
output = output.append(africa_df, ignore_index = True)
output = output.append(oceania_df, ignore_index = True)
output = output.append(eur_df, ignore_index = True)
output
```

```python
output.to_csv('../data/preprocessed_data.csv', index=False)
```
